```python
from yolov3.utils import *
import cv2
from matplotlib import pyplot as plt
image_path   = "/content/gdrive/MyDrive/YOLOV3teststage/006525.jpg"
result = detect_image(yolo, image_path, "/content/gdrive/MyDrive/YOLOV3teststage/newpic.jpg", input_size=YOLO_INPUT_SIZE, show=False, CLASSES=TRAIN_CLASSES, rectangle_colors=(255,0,0))
#print(result[1])
#image=result[0]
image = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(30,15))
plt.imshow(image)
```
